```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
car_crashes = sns.load_dataset("car_crashes")
```

```python
car_crashes
```

```python
print(car_crashes.head())
```

```python
sns.distplot(car_crashes["alcohol"])
plt.show()
```

```python
sns.distplot(car_crashes["alcohol"], kde = False)
plt.show()
```

```python
sns.distplot(car_crashes["alcohol"], hist = False)
plt.show()
```

```python
sns.jointplot(x="speeding", y="alcohol", data=car_crashes)
plt.show()
```

```python
sns.set_style("ticks")
sns.pairplot(car_crashes)
plt.show()
```

```python
sns.stripplot(x="abbrev", y="speeding", data=car_crashes)
plt.show()
```

```python
sns.stripplot(x="abbrev", y="speeding", data=car_crashes, jitter=True)
plt.show()
```

```python
sns.swarmplot(x="abbrev", y="speeding", data=car_crashes)
plt.show()
```

```python
sns.boxplot(x=car_crashes["alcohol"])
plt.show()
```

```python
sns.boxplot(x="abbrev", y="alcohol", data=car_crashes)
plt.show()
```

```python
sns.violinplot(x=car_crashes["speeding"])
plt.show()
```

```python
sns.violinplot(x="abbrev", y="speeding", data=car_crashes)
plt.show()
```

```python
sns.barplot(x="abbrev", y="alcohol", data=car_crashes)
plt.show()
```

```python
sns.regplot(x="total", y="alcohol", data=car_crashes)
plt.show()
```

```python
sns.lmplot(x="total", y="alcohol", data=car_crashes)
plt.show()
```

```python
sns.residplot(x="total", y="alcohol", data=car_crashes)
plt.show()
```

```python
uniform_data = np.random.rand(10, 30)
sns.heatmap(uniform_data)
plt.show()
```

```python
subset1 = car_crashes[['total', 'speeding', 'alcohol']]
sns.clustermap(subset1)
plt.show()
```

```python
g = sns.FacetGrid(car_crashes, col="speeding", row="alcohol")
g = g.map(plt.hist, "total")
plt.show()
```

```python
g = sns.JointGrid(x="total", y="alcohol", data=car_crashes)

g = g.plot(sns.regplot, sns.distplot)
```
